```python
# Import libraries
html_tables = {}

# For every table in the datasets folder...
for table_name in os.listdir('data'):
    #this is the path to the file
    table_path = f'data/{table_name}'
    # Open as a python file in read-only mode
    table_file = open(table_path,"r")
    # Read the contents of the file into 'html'
    html_temp = table_file.read()
    html = BeautifulSoup(html_temp, "html.parser")
    # Find 'news-table' in the Soup and load it into 'html_table'
    html_table = html.find("table",{"id":"news-table"})
    print(html_table)
    # Add the table to our dictionary
    html_tables[table_name] = html_table
```

```python
meta = html_tables['meta.html']

rows = meta.find_all('tr')

print(rows)
```

```python
for i, table_row in enumerate(rows):
    # Read the text of the element 'a' into 'link_text'
    link_text = table_row.find_all('a')
    # Read the text of the element 'td' into 'data_text'
    data_text = table_row.find_all('td')
    # Print the count
    print(f'File number {i+1}:')
    # Print the contents of 'link_text' and 'data_text'
    print(link_text)
    print(data_text)
    # The following exits the loop after four rows to prevent spamming the notebook, do not touch
    if i == 3:
        break
```
